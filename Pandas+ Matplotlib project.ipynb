Pandas + Matplotlib Project

Accept the class1 percentage class2, class3 as per user

```python
import pandas as pd
num = int(input("How many students are there? "))
name = []
class1_per = []
class2_per = []
class3_per = []

for i in range(num):
    nm = input("Enter the student name - ")
    c1 = float(input("Enter the class 1 percentage - "))
    c2 = float(input("Enter the class 2 percentage - "))
    c3 = float(input("Enter the class 3 percentage - "))
    name.append(nm)
    class1_per.append(c1)
    class2_per.append(c2)
    class3_per.append(c3)
   
```

Create a DataFrame out of this Data out of this

```python
data = {"Student Name": name , "Class1 Percentage" : class1_per , "Class2 Percentage" : class2_per , "Class3 Percentage" : class3_per}
df = pd.DataFrame(data)
print(df)

```

Get values from the columns of the DataFrame and plot a Comparison Plot (plot every data in single plot) With different Colors

```python
import matplotlib.pyplot as plt

plt.style.use("ggplot")
plt.figure(figsize = (8,5))

df.plot(x = "Student Name", y = ["Class1 Percentage", "Class2 Percentage", "Class3 Percentage"], kind = "bar")

plt.title("Percentage of Students", color = "red")

plt.xlabel("Student name", color = "red")
plt.ylabel("Percentage" , color = "red")

plt.bar(x,y, color = "cyan")
plt.show()
```
